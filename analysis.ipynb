```python
import seaborn as sns
import pandas as pd
from scipy.stats import ks_2samp
```

```python
vars = ["A","B","C","D","E","F"]
```

```python
def analysis(path):
    dataframe = pd.read_csv(path, usecols= vars)
    means = round(dataframe.mean(axis=0),2)
    variances = round(dataframe.var(axis=0),2)
    corrs = round(dataframe.corr(),3)
    return dataframe, means, variances, corrs
```

```python
def ks_2(df1, df2, vars, skip):
    alpha_bonf = 0.05/(len(vars)-1)
    print("Alpha_bonferoni correction:", alpha_bonf)
    
    for var in vars:
        if var != skip:
            result = ks_2samp(df1[var], df2[var])
            pvalue = result.pvalue
            if pvalue < alpha_bonf:
                print(f"{var} vs {var}: Null hypothesis is rejected with p-value {round(pvalue,4)}")
            else:
                print(f"{var} vs {var}: Null hypothesis can't be rejected with p-value {round(pvalue,4)}")
```
