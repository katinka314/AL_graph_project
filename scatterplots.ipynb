```python
import numpy as np

def condition_on_interval(data, idx, low, high):
    mask = (data[:, idx] >= low) & (data[:, idx] <= high)
    return data[mask]
```

```python
cond_base = condition_on_interval(basedata, 2, 0, 2)
plot_pairwise_hist(basedata, cond_base)
```

```python
import numpy as np

def remove_top_x(arr, x):
    arr_sorted = np.sort(arr)          # ascending
    return arr_sorted[:-x]             # remove largest x values

# example
clean = remove_top_x(basedata[:,1], x=2)

plt.hist(clean)
plt.show()
```
