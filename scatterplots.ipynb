```python
import numpy as np

def condition_on_interval(data, idx, low, high):
    mask = (data[:, idx] >= low) & (data[:, idx] <= high)
    return data[mask]
```

```python
cond_base = condition_on_interval(basedata, 5, 0, 5)
plot_grid(cond_base, n=6)
```

```python
cond_base = condition_on_interval(basedata, 5, -5, 0)
plot_grid(cond_base, n=6)
```
